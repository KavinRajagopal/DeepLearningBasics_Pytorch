```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym #sympy as partial derivative

from IPython import display 
display.set_matplotlib_formats('svg')
```

```python
  #the peaks of the function
def peaks(x,y):
    x,y = np.meshgrid(x,y)
 
    z = 3 *(1-x)**2 * np.exp(-(x**2) - (y+1)**2) \
      - 10 * (x/5 - x**3 - y**5) * np.exp(-x**2 - y**2) \
      - 1/3 * np.exp(-(x+1)**2 - y **2)


```

```python
#create the landscape
x = np.linspace(-3,3, 201)
y = np.linspace(-3,3, 201)


Z = peaks(x,y)

plt.imshow(Z, extent = [x[0], x[-1], y[0], y[-1]], vmin = -5, vmax = 5, origin = 'lower', dtype = 'float')
plt.show()
```
