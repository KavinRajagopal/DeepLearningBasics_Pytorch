```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

Running experiments to understand gradient descent 


```python
x = np.linspace(-2*np.pi, 2*np.pi, 401)
fx = np.sin(x) * np.exp(-x**2*0.05)


df = np.cos(x) * np.exp(-x**2*0.05) + np.exp(-x**2*0.05) * (- .1 * x) * np.exp(-x**2*0.05)


#plotting the graph 



plt.plot(x, fx, x, df)
plt.legend(['f(x)', 'df'])
```

```python
def fx(x):
  return np.sin(x) * np.exp(-x**2*0.05)

def deriv(x):
   return np.cos(x) * np.exp(-x**2*0.05) + np.exp(-x**2*0.05) * (- .1 * x) * np.exp(-x**2*0.05)

```

```python
localmin = np.random.choice(x,1)


#setting parameters
learning_rate = 0.01
training_epochs = 100

for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin = learning_rate * grad


plt.plot(x, fx(x), x, deriv(x), '--')
plt.plot(localmin, deriv(localmin), 'ro')
plt.plot(localmin, fx(localmin), 'ro')


plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x', 'df', 'f(x) min'])
plt.title('Empirical local minimum: %s ' %localmin[0])
plt.show()
```
